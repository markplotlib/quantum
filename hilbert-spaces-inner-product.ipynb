# Hilbert Spaces

What is a **Hilbert Space**?

You can think of it as:

> the state space in which all quantum state vectors *live*.

[- Qiskit](https://qiskit.org/textbook/ch-appendix/linear_algebra.html)

Recall that a quantum state vector is a qubit (represented by a vector), and a quantum gate enables a quantum computation (represented by a matrix).

What differentiates a *Hilbert Space* from other vector spaces?

Hilbert Spaces are equipped with an **inner product**: an operation performed between two vectors that returns a scalar.

This scalar represents a *projection* of the first vector onto the second.

```python
import numpy as np
```

Inner product: **`<a|b>`**.

Note: `<a|` is the *conjugate transpose* of `|a>`, denoted by 
`|a⟩†`.

A vector's inner product with itself is 1: `<ψ|ψ>` = 1.
This is the *normalization condition*.

```python
a = np.array([3/5, 4/5], dtype=float)

np.matmul(a, a.conjugate().transpose())
```

```python
# vector matrices need no transposing in NumPy
np.matmul(a, a)
```

If they are orthogonal, the inner product is zero.

```python
b = np.array([-4/5, 3/5])

np.matmul(a, b)
```

## unitary matrices and Hilbert spaces

#### Unitary matrices *preserve the inner product*.

> No matter how you transform a vector under a sequence of unitary matrices, the normalization condition still holds true.

```python
from matplotlib import pyplot as plt
from qiskit import *
from qiskit.visualization import plot_bloch_vector
```

```python
plot_bloch_vector([0, 0, 0])
```

The surface of this sphere is a valid Hilbert space, as is the inner product of qubit state vectors.

```python
c = np.array([0, 1], dtype=float)
d = np.array([1, 0], dtype=float)

np.matmul(c, d)
```

```python
from math import sqrt
 
e = np.array([1/sqrt(2), 1/sqrt(2)], dtype=float)

np.matmul(c, e)
```

```python
round(np.matmul(e, e), 2)
```
